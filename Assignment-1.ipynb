```python
data.rename(columns={
    'Pclass': 'PassengerClass',
    'Fare': 'TicketFare',
    'Sex': 'Gender'
}, inplace=True)
```

```python
print(data.columns)
```

```python
data['Survived'] = data['Survived'].astype(str)
```

```python
print(data.dtypes)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 2: Now plot
sns.countplot(x='Survived', data=data)
plt.title("Number of Survivors (0 = No, 1 = Yes)")
plt.show()

```

```python
sns.countplot(x='Gender', data=data)
plt.title("Gender Distribution")
plt.show()
```

```python
sns.countplot(x='Gender', data=data)
plt.title("Gender Distribution")
```

```python
data['Embarked'].value_counts().plot.pie(autopct='%1.1f%%', figsize=(6,6))
plt.title("Port of Embarkation")
plt.ylabel('')
plt.show()
```

```python

# Plot pie chart for Passenger Class
data['Pclass'].value_counts().sort_index().plot.pie(
    autopct='%1.1f%%',
    labels=['Class 1', 'Class 2', 'Class 3'],
    startangle=90
)

plt.title("Passenger Class Distribution")
plt.ylabel('')
plt.show()
```

```python
import matplotlib.pyplot as plt

data['Age'].plot.hist(bins=10, color='pink', edgecolor='black')

plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Number of Passengers")
plt.show()
```

```python
sns.scatterplot(x='Age', y='TicketFare', hue='Survived', data=data)
plt.title("Age vs Fare (Colored by Survival)")
plt.show()
```
